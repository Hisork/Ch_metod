**Вопрос 1:** Как можно оценить качество модели?

**Ответ:** Качество модели можно оценивать при помощи следующих критериев:

1.   Сумма квадратов ошибок отклонений
2.   Сумма модулей ошибок отклонений
3.   Минимум максимальной по модулю ошибки.

Для оценки нашей модели будет использован первый критерий



**Вопрос 2:** Постройте график остатков (разность между исходными данными и моделью). Первоначально построим заданную модель:

```python
N = 100    # число наблюдений (выборка данных)
k = 2    # теоретическое значение параметра k
b = 3       # теоретическое значение параметра b
d =4  # диапазон значений x
n=0
```

```python
sigma = 1
x = np.array(range(N))/N*d-d/2
f = np.array([k*z+b for z in x])
y = f + np.random.normal(0, sigma, N)

plt.plot(x,f,c='red')
plt.scatter(x,y,s=3,c='green')
plt.grid(True)
plt.show()
```

```python
mx = x.sum()/N
my = y.sum()/N
a2 = np.dot(x.T, x)/N
a11 = np.dot(x.T, y)/N

kn = (a11 - mx*my)/(a2 - mx**2)
bn = my - kn*mx
fn = np.array([kn*z+bn for z in x])

plt.plot(x,f,c='red')
plt.scatter(x,y,s=3,c='green')
plt.grid(True)
plt.plot(x,fn)
plt.show()
```

Далее строим график остатков, путем получения разности между исходными данными и моделью:

```python
plt.plot(x,f-fn,c='red')
plt.grid(True)
plt.show()
```

График так же отображает четкую линейную зависимость, что говорит о высоком качестве построенной нами модели.

**Вопрос 3:** Рассчитайте стандартное отклонение остатков sfn.

Стандартное отклонение вычисляется путем рассчета квадратного корня от от сумм квадратов отклонений, деленных на количество измерений минус порядок уравнения (в нашем случае это N-2, т.е. 98).


```python
sfn=((sum(f-fn)/(N-2)*sum(f-fn)/(N-2)))**(1/2)
print(sfn)
```

```python
sfn=((sum(f-fn)/(N-2)*sum(f-fn)/(N-2)))**(1/2)
print(sfn)
```

**Вопрос 6:** Постройте зависимость sfn от сигмы.

**Ответ:** 

```python
N = 10    # число наблюдений (выборка данных)
sigma = 1
k = 2    # теоретическое значение параметра k
b = 3       # теоретическое значение параметра b
d =4  # диапазон значений x
n=0
sigma_arr=[]
sfn_arr=[]
while (n<10):
    sigma = n# стандартное отклонение (разброс) наблюдаемых значений
    sigma_arr.append(sigma)
    n=n+0.5
    x = np.array(range(N))/N*d-d/2
    f = np.array([k*z+b for z in x])
    y = f + np.random.normal(0, sigma, N)
    mx = x.sum()/N
    my = y.sum()/N
    a2 = np.dot(x.T, x)/N
    a11 = np.dot(x.T, y)/N
    
    kn = (a11 - mx*my)/(a2 - mx**2)
    bn = my - kn*mx
    fn = np.array([kn*z+bn for z in x])

    sfn=((sum(f-fn)/(N-2)*sum(f-fn)/(N-2)))**(1/2)
    sfn_arr.append(sfn)
plt.plot()
plt.scatter(sigma_arr,sfn_arr,c='red')
plt.show

```

**Вопрос 7:** Постройте зависимость sfn от размера выборки данных (10,200).

Вопрос 8: Реализовать МНК для параболы

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x = np.linspace(-10, 10, 21)  # 21 points evenly spaced between -10 and 10
y = 2 * x**2 + 3 * x - 1 + np.random.normal(scale=5, size=x.shape)  # parabola with noise

def parabola(x, a, b, c):
    return a * x**2 + b * x + c

params, _ = curve_fit(parabola, x, y)

plt.plot(x, y, 'o')
plt.plot(x, parabola(x, *params))
plt.show()
```

В этом примере мы сначала генерируем некоторые выборочные данные, состоящие из 21 точки с добавлением шума. Затем мы определяем функцию параболы для нашей модели с тремя параметрами a, b и c, представляющими коэффициенты квадратного уравнения. Используем функцию curve_fit из модуля scipy.optimize, чтобы найти значения этих параметров, которые минимизируют сумму квадратов различий между прогнозируемыми значениями модели и фактическими точками данных.
